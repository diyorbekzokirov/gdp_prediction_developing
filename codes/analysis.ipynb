```python
# Identify Country Codes in each dataframe
codes_df1 = set(df1['Country Code'].dropna().unique())
codes_df2 = set(df2['Country Code'].dropna().unique())
codes_df3 = set(df3['Country Code'].dropna().unique())
codes_df4 = set(df4['Country Code'].dropna().unique())
codes_df5 = set(df5['Country Code'].dropna().unique())
codes_df6 = set(df6['Country Code'].dropna().unique())
codes_df7 = set(df7['Country Code'].dropna().unique())
codes_df8 = set(df8['Country Code'].dropna().unique())

print("Original Country Codes:")
print(f"df1: {len(codes_df1)} countries")
print(f"df2: {len(codes_df2)} countries")
print(f"df3: {len(codes_df3)} countries")
print(f"df4: {len(codes_df4)} countries")
print(f"df5: {len(codes_df5)} countries")
print(f"df6: {len(codes_df6)} countries")
print(f"df7: {len(codes_df7)} countries")
print(f"df8: {len(codes_df8)} countries")

# Find duplicates across datasets
duplicates_in_df2 = codes_df2 & codes_df1
duplicates_in_df3 = codes_df3 & (codes_df1 | codes_df2)
duplicates_in_df4 = codes_df4 & (codes_df1 | codes_df2 | codes_df3)
duplicates_in_df5 = codes_df5 & (codes_df1 | codes_df2 | codes_df3 | codes_df4)
duplicates_in_df6 = codes_df6 & (codes_df1 | codes_df2 | codes_df3 | codes_df4 | codes_df5)
duplicates_in_df7 = codes_df7 & (codes_df1 | codes_df2 | codes_df3 | codes_df4 | codes_df5 | codes_df6)
duplicates_in_df8 = codes_df8 & (codes_df1 | codes_df2 | codes_df3 | codes_df4 | codes_df5 | codes_df6 | codes_df7)

print(f"\nDuplicates found:")
print(f"df2 has duplicates with df1: {duplicates_in_df2}")
print(f"df3 has duplicates with df1/df2: {duplicates_in_df3}")
print(f"df4 has duplicates with df1/df2/df3: {duplicates_in_df4}")
print(f"df5 has duplicates with df1/df2/df3/df4: {duplicates_in_df5}")
print(f"df6 has duplicates with df1/df2/df3/df4/df5: {duplicates_in_df6}")
print(f"df7 has duplicates with df1/df2/df3/df4/df5/df6: {duplicates_in_df7}")
print(f"df8 has duplicates with df1/df2/df3/df4/df5/df6/df7: {duplicates_in_df8}")

# Remove duplicates: keep in first dataset, drop from others
df2_cleaned = df2[~df2['Country Code'].isin(duplicates_in_df2)]
df3_cleaned = df3[~df3['Country Code'].isin(duplicates_in_df3)]
df4_cleaned = df4[~df4['Country Code'].isin(duplicates_in_df4)]
df5_cleaned = df5[~df5['Country Code'].isin(duplicates_in_df5)]
df6_cleaned = df6[~df6['Country Code'].isin(duplicates_in_df6)]
df7_cleaned = df7[~df7['Country Code'].isin(duplicates_in_df7)]
df8_cleaned = df8[~df8['Country Code'].isin(duplicates_in_df8)]

print(f"\nAfter removing duplicates:")
print(f"df1: {len(df1)} rows")
print(f"df2_cleaned: {len(df2_cleaned)} rows (removed {len(df2) - len(df2_cleaned)})")
print(f"df3_cleaned: {len(df3_cleaned)} rows (removed {len(df3) - len(df3_cleaned)})")
print(f"df4_cleaned: {len(df4_cleaned)} rows (removed {len(df4) - len(df4_cleaned)})")
print(f"df5_cleaned: {len(df5_cleaned)} rows (removed {len(df5) - len(df5_cleaned)})")
print(f"df6_cleaned: {len(df6_cleaned)} rows (removed {len(df6) - len(df6_cleaned)})")
print(f"df7_cleaned: {len(df7_cleaned)} rows (removed {len(df7) - len(df7_cleaned)})")
print(f"df8_cleaned: {len(df8_cleaned)} rows (removed {len(df8) - len(df8_cleaned)})")

# Merge all cleaned dataframes
merged_df = pd.concat([df1, df2_cleaned, df3_cleaned, df4_cleaned, df5_cleaned, df6_cleaned, df7_cleaned, df8_cleaned], ignore_index=True)

print(f"\nFinal merged dataframe:")
print(f"Total rows: {len(merged_df)}")
print(f"Total unique Country Codes: {len(merged_df['Country Code'].dropna().unique())}")
print(f"Shape: {merged_df.shape}")
```

```python
id_cols = ['Country Name', 'Country Code', 'Series Name', 'Series Code']

# Year columns (format: YYYY [YRyyyy])
year_cols = [col for col in merged_df.columns if col not in id_cols]

# Unpivot the dataframe
transformed_df = merged_df.melt(
    id_vars=id_cols,
    value_vars=year_cols,
    var_name='Year',
    value_name='Value'
)

# Extract the year number from the Year column (e.g., "1960 [YR1960]" -> 1960)
transformed_df['Year'] = transformed_df['Year'].str.extract(r'(\d{4})')[0].astype(int)

# Pivot so that Series Name becomes columns and we have Country Code, Year, and all Series as columns
final_df = transformed_df.pivot_table(
    index=['Country Code', 'Country Name', 'Year'],
    columns='Series Name',
    values='Value',
    aggfunc='first'
).reset_index()

print(f"Final dataframe shape: {final_df.shape}")
print(f"Columns: {final_df.columns.tolist()}")
print(f"\nFirst few rows:")
print(final_df.head(10))

print(f"\nNumber of Series (columns): {len(final_df.columns) - 3}")  # -3 for Country Code, Country Name, Year
```
